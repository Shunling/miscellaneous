```python
from Bio import Entrez
import requests, bs4
import json
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('test2.csv') # read the download table from ClinVar under search multiple sclerosis
df.head()
```

```python
IDlist = [str(int(x[4:])) for x in df.Accession.values] # extract ClinVar accession ID
```

```python
geneID = [] # store geneID
```

```python
for i, Id in enumerate(IDlist): # find full infomation of link
    urlstem = 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi?db=clinvar&retmode=xml&id='
    url = urlstem + Id 
    r = requests.get(url) # requrest full web information
    
    record = re.findall(r'<GeneID>(.*)</GeneID>', str(r.content)) # extract gene ID
    geneID.append(record)
    print(Id, record) # track abnormalities
```

```python
df['geneIDs'] = geneID[-1972:] # drop the first accidentally curated Ids, and add geneID column to data
```
