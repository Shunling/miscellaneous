```python
# directrly parse GSE file from GSE ID.
gse = GEOparse.get_GEO(geo='GSE24742',destdir='./')
```

```python
for gpl_name, gpl in gse.gpls.items(): # platform and metadata
    print('Name:',gpl_name) # platform name
    print('metadata:')
    for key, value in gpl.metadata.items(): # print metadata
        print(' - %s : %s' % (key, ','.join(value)))
    print('table data:',)
    print(gpl.table.head()) # print first 5 table content 
    
```
